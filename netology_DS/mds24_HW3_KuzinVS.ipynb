```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

##Задание 1
Дан объект в  пространстве

```python
A = [
  [-100, -100, 1],
  [-100, 100, 1],
  [100, 100, 1],
  [0, 0, 1],
  [100, -100, 1],
  [-100, -100, 1]]

A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

При помощи линейных отображений:

* Уменьшить объект в два раза
* Повернуть на 130 градусов
* Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])

```python
x2 = x / 2
y2 = y / 2

plt.plot(x2, y2)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Повернуть на 130 градусов

degr = np.radians(130)
turn = np.array([[np.cos(degr),np.sin(degr), 0], [-np.sin(degr),np.cos(degr), 0], [0,0,1]])

a_turn = A.dot(turn)
x_turn = a_turn[:,0]
y_turn = a_turn[:,1]
plt.plot(x_turn, y_turn)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])
rvrs = [[0, 1, 0], [1, 0, 0], [0, 0, 1]]
a_rvrs = A.dot(rvrs)
x_rvrs = a_rvrs[:,0]
y_rvrs = a_rvrs[:,1]
plt.plot(x_rvrs, y_rvrs)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```
